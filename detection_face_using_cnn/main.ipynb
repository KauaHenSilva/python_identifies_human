```python
import cv2
import dlib
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Lena.jpg')

imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

detec = dlib.cnn_face_detection_model_v1('mmod_human_face_detector.dat')
facesDetected = detec(imgGray, 1)

for face in facesDetected:
    x, y, w, h, c = face.rect.left(), face.rect.top(), face.rect.right(), face.rect.bottom(), face.confidence
    cv2.rectangle(imgRGB, (x, y), (w, h), (255, 0, 0), 3)
    
plt.imshow(imgRGB)
```
